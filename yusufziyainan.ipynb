```python
# Günlük Zaman Değerlendirme Programı

def gunluk_zaman_al():
    while True:
        try:
            uyku = float(input("Günlük uyku süresini (saat) giriniz: "))
            ders = float(input("Günlük ders süresini (saat) giriniz: "))
            serbest = float(input("Günlük serbest zaman süresini (saat) giriniz: "))

            # Negatif kontrolü
            if uyku < 0 or ders < 0 or serbest < 0:
                print("Hata: Negatif değer girilemez.\n")
                continue

            # Toplam saat kontrolü
            if uyku + ders + serbest > 24:
                print("Hata: Sürelerin toplamı 24 saati geçemez.\n")
                continue

            return uyku, ders, serbest

        except ValueError:
            print("Hata: Lütfen sayısal bir değer giriniz.\n")


def uyku_durumu_belirle(uyku):
    if uyku < 6:
        return "Yetersiz uyku"
    elif 6 <= uyku <= 8:
        return "Normal uyku"
    else:
        return "Fazla uyku"


def ders_durumu_belirle(ders):
    if ders >= 6:
        return "Yoğun çalışma günü"
    else:
        return "Normal çalışma günü"


def genel_durum_belirle(uyku, ders):
    if uyku < 6 and ders >= 6:
        return "Verimsiz gün"
    else:
        return "Dengeli gün"


# Ana program
uyku, ders, serbest = gunluk_zaman_al()

uyku_durumu = uyku_durumu_belirle(uyku)
ders_durumu = ders_durumu_belirle(ders)
genel_durum = genel_durum_belirle(uyku, ders)

print("\n--- GÜNLÜK DEĞERLENDİRME ---")
print("Uyku durumu:", uyku_durumu)
print("Ders durumu:", ders_durumu)
print("Genel değerlendirme:", genel_durum)
```
